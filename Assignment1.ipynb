```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
# plt.style.use('seaborn-whitegrid')
```

```python
df = pd.read_csv('/Users/cicipan/projects/CS448B/simpsons_episodes.csv')
```

```python
df_f = df.drop(labels=['id','image_url','production_code','video_url'], axis=1)
```

```python
df_f
```

```python
df_f.describe(include=np.number)
```

```python
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 30})
plt.style.use('default')
plt.scatter(x=df['original_air_year'], y=df['views'], c=df['imdb_rating'], cmap='viridis',
                  alpha=0.4, s=(df_f['us_viewers_in_millions']-2)*10, label=None)
plt.xlabel('Original Year')
plt.ylabel('Online Views')
plt.colorbar().ax.set_ylabel('IMDb Rating')

# draw average line for online views
y=df['views']
plt.axhline(y.mean(), color='r', linestyle='dashed', linewidth=1)

# Here we create a legend:
# we'll plot empty lists with the desired size and label
for area in [3, 10, 20, 25, 33]:
    plt.scatter([],[], c='k', alpha=0.2, s=(area-2)*10,
                label=str(area) + ' millions')

plt.legend(scatterpoints=1, labelspacing=1, title='Total US Views')

plt.title('Simpsons Online Views and IMDb Rating')

plt.savefig('Assignment1.png')
```
